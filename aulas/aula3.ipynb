### Adicionar uma coluna 

- A partir de uma coluna que existe

```python
import pandas as pd
```

```python
vendas_df = pd.read_excel('Vendas.xlsx')
```

```python
display(vendas_df)
```

```python
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
```

```python
display(vendas_df)
```

_Nesse caso, pegamos a coluna 'Valor Final' que já existia e alteramos para 'Comissão', e o seu valor foi multiplicado por 0.05 que corresponde a 5% do valor final._

* Criar uma coluna com valor padrão

```python
vendas_df.loc[:, 'Imposto'] = 0
```

```python
display(vendas_df)
```

_Selecionamos todas as linhas utilizando dois pontos (:), criamos uma nova coluna chamada 'Imposto' e atribuimos o valor 0 para todas as linhas._

### Adicionar uma linha

_Linhas de um complemento da base de dados_

```python
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
# display(vendas_dez_df)
```

```python
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

### Excluir linhas e colunas

~> Excluindo uma coluna

```python
vendas_df = vendas_df.drop('Imposto', axis=1)
display(vendas_df)
```

_A coluna 'Imposto' foi excluída. É importante lembrar que sempre temos que passar o eixo (axis), 0 para linha e 1 para coluna._

~> Excluindo uma linha

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

_A linha (índice) 0 foi excluída._

### Tratando valores vazios

- Deletar linhas/colunas vazias
- Deletar linhas que possuem valores vazios(NaN)
- Preencher valores vazios (média e último valor)

1 - Deletar linhas e colunas completamente vazias

```python
vendas_df = vendas_df.dropna(how='all')
```

_dropna(how='all') -> o comando dropna com o parâmetro how='all' serve para excluir linhas e colunas completamente vazias, e em caso de exclusão específica como, excluir apenas colunas OU linhas vazias, passamos o parâmetro axis=0 ou axis=1._

2 - Deletar linhas que possuem pelo menos um valor vazio

```python
vendas_df = vendas_df.dropna()
```

_Passamos apenas o dropna vazio, que é o formato padrão para excluir todas as linhas que possuem pelo menos um valor vazio._

3 - Preencher valores vazios

- _Preencher com a média da coluna_

```python
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)
```

_Utilizamos o fillna() para preencher os valores vazios com um valor específico. Nesse modelo, preenchemos com a média da coluna 'Comissão'._

* _Preencher com o último valor_

```python
vendas_df = vendas_df.ffill()
```

_O ffill() vai preencher com o primeiro valor acima dele._

### Calcular indicadores

* Groupby

```python
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

_O método groupby() permite agrupar dados por uma ou mais colunas e aplicar funções de agregação (como sum(), mean(), count() etc.) aos grupos formados._

_É muito útil para resumir dados agrupando-os por categorias ou características específicas e calculando estatísticas para cada grupo._

_Nesse exemplo, usamos groupby('Produto') para agrupar os dados pelo produto e, em seguida, calculamos a soma do valor final de cada produto._

* Value Counts

```python
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

_O método value_counts() conta a frequência de valores únicos em uma coluna específica._

_Esse método é útil para entender a distribuição de uma variável categórica ou identificar a quantidade de ocorrências de cada valor._

_Nesse exemplo, usamos o value_counts() para calcular a quantidade de transações realizadas em cada loja._

### Mesclar dois Dataframes (Procurar informações de um Dataframe em outro)

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
# display(gerentes_df)

vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```

_Utilizamos o método merge para mesclar os Dataframes._
